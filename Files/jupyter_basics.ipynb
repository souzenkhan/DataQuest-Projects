```python
import csv

with open('/Users/souzenkhan/Downloads/yelp_dataset/yelp_business_restaurants.csv', 'r') as file:
    reader = csv.reader(file)
    df = list(reader)
```

```python
'''
Function to explore data in a DataFrame (DataQuest written function)
'''
def explore_data(df, start, end, rows_and_columns=False):
    dataset_slice = df[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(df))
        print('Number of columns:', len(df[0]))

```

```python
explore_data(df, 0, 5, True)
```
